# USA State Poll Results for the 2012 & 2016 Presidential Elections

This script scrapes the wikipedia pages for the USA 2012 and 2016 presidential election state polls and saves the results in 2 separate csv files

```python
### Libraries
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
poll_states = ["Alabama",
               "Alaska",
               "Arizona",
               "Arkansas",   
               "California",
               "Colorado",
               "Connecticut",
               "Delaware",
               "District of Columbia",
               "Florida",
               "Georgia",
               "Hawaii",
               "Idaho",
               "Illinois",
               "Indiana",
               "Iowa",
               "Kansas",
               "Kentucky",
               "Louisiana",
               "Maine",
               "Maryland",
               "Massachusetts",
               "Michigan",
               "Minnesota",
               "Mississippi",
               "Missouri",
               "Montana",
               "Nebraska",
               "Nevada",
               "New Hampshire",
               "New Jersey",
               "New Mexico",
               "New York",
               "North Carolina",
               "North Dakota",
               "Ohio",
               "Oklahoma",
               "Oregon",
               "Pennsylvania",
               "Rhode Island",
               "South Carolina",
               "South Dakota",
               "Tennessee",
               "Texas",
               "Utah",
               "Vermont",
               "Virginia",
               "Washington",
               "West Virginia",
               "Wisconsin",
               "Wyoming"]
```

We loop through the tables and extract the relevant information from each row and also add the state that the polls are from (the states are not in the table so we have to add them manually). The main idea of the scraping comes from https://adesquared.wordpress.com/2013/06/16/using-python-beautifulsoup-to-scrape-a-wikipedia-table/. Note: Tenessee's 8th entry is buggy, so have to correct manually after scraping.

```python
### Extract HTML using BeautifulSoup
html = urlopen("https://en.wikipedia.org/wiki/Statewide_opinion_polling_for_the_United_States_presidential_election,_2008")
bsObj = BeautifulSoup(html.read())

### Declare variables
poll_source=""
date=""
dem_perc=""
rep_perc=""

### Loop through tables, extract relevant information save them in a CSV file
f = open('polls2008.csv', 'w')
i=0 #to keep track of current state
for table in bsObj.find_all("table", { "class" : "wikitable" }):
    if (i<len(poll_states)):
        state=poll_states[i]
        i=i+1
        for rows in table.find_all("tr"):
            row = rows.find_all("td")
            if (len(row)==7 or len(row)==9): #to include those that have 3 way races. 1 (Virginia) also has a 4 way race, but no sample size or error so it has the same amount of cells as 3
                poll_source=row[0].find(text=True)
                date=row[1].find(text=True)
                dem_perc=row[3].find(text=True)
                rep_source=row[5].find(text=True)
                write_to_file = state + "," + poll_source + "," + date + "," + dem_perc + "," + rep_source + "\n" 
                print(write_to_file)
                f.write(write_to_file)
f.close()
```

## 2016

Again we e loop through the tables and extract the relevant information from each row and also add the state that the polls are from (the states are not in the table so we have to add them manually). We also skip the first table (the national polls) and for every state, we also skip the first table (which are the aggregate polls). We do both of these through the if statement in the text. **Note**: Oregon's first entry and Pennsylvania's 25th and 41st entry are buggy so you have to adjust it manually after you run the script (the poll name has ',' in it so the rows get messed up)

```python
### Extract HTML using BeautifulSoup
html = urlopen("https://en.wikipedia.org/wiki/Statewide_opinion_polling_for_the_United_States_presidential_election,_2016")
bsObj = BeautifulSoup(html.read())

### Delcare variables
poll_source=""
date=""
dem_perc=""
rep_perc=""

### Loop through tables, extract relevant information save them in a CSV file
f = open('polls2016.csv', 'w')
i=0 #to keep track of state
for table in bsObj.find_all("table", { "class" : "wikitable" }):
    state=poll_states[i]
    if (i is not 0 and state is not ""): #"0" is the national polls and every "" corresponds to the aggregate polls
        for rows in table.find_all("tr"):
            row = rows.find_all("td")
            if (len(row)==9 or len(row)==11 or len(row)==13 or len(row)==15 or len(row)==17): #to include those that have 3, 4, 5 & 6 way races.
                poll_source=row[0].find(text=True)
            #print(poll_source)
                date=row[1].find(text=True)
            #print(date)
                dem_perc=row[3].find(text=True)
            #print(dem_perc)
                rep_source=row[5].find(text=True)
            #print(rep_source)
                write_to_file = state + "," + poll_source + "," + date + "," + dem_perc + "," + rep_source + "\n" 
                print(write_to_file)
                f.write(write_to_file)
    i=i+1
f.close()
```
